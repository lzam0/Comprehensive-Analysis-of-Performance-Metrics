```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Attempt to read the CSV files and handle exceptions
df_ucla = pd.read_csv("ucla_2023.csv", encoding='latin1')
df_hawaii = pd.read_csv('hawaii_2023.csv', encoding='latin1')

df_hawaii.fillna(0, inplace = True) # replace all empty cells in DF with 0
df_ucla.fillna(0, inplace = True) # replace all empty cells in DF with 0

# print(df_hawaii, df_ucla)
```

```python
# print height of every player inside of team hawaii with name and position
hawaii_player_details = df_hawaii[['Player', 'Pos','Ht']]
ucla_player_details = df_ucla[['Player', 'Pos','Ht']]
# print(hawaii_player_details, ucla_player_details)
```

```python
# Convert ft to inches and then to centimeters for UCLA
df_ucla['Height'] = df_ucla['Ht'].apply(lambda x: (int(x.split("'")[0]) * 12 + int(x.split("'")[1].split('"')[0])) * 2.54)

# Sort UCLA DataFrame by height
df_ucla_height_sorted = df_ucla.sort_values(by='Height')

# Convert ft to inches and then to centimeters for Hawaii
df_hawaii['Height'] = df_hawaii['Ht'].apply(lambda x: (int(x.split("'")[0]) * 12 + int(x.split("'")[1].split('"')[0])) * 2.54)

# Sort Hawaii DataFrame by height
df_hawaii_height_sorted = df_hawaii.sort_values(by='Height')

# Calculate average height for each position in UCLA
ucla_avg_heights = df_ucla_height_sorted.groupby('Pos')['Height'].mean()

# Calculate average height for each position in Hawaii
hawaii_avg_heights = df_hawaii_height_sorted.groupby('Pos')['Height'].mean()
```

```python
# Plot
pos_ucla = ucla_avg_heights.index  # Assuming both teams have the same positions
pos_hawaii = hawaii_avg_heights.index  # Assuming both teams have the same positions


plt.figure(figsize=(8, 6))

plt.bar(pos_ucla, ucla_avg_heights, color='skyblue', label='UCLA', alpha=1)

plt.title("Height of UCLA Volleyball Players 2023")
plt.xlabel('Position')
plt.ylabel('Average Height (in cm)')
plt.legend()
plt.show()
```

What does this indicate to us?

A dig is defined as a controlled pass of a hard driven ball coming from the opponent's attack, usually a spike or a serve

- UCLA Dig Total: 907.0
- Hawaii Dig Total: 944.0
- Total Digs: 1851.0

- Hawaii is 1.99892% has a better defence than UCLA

The result in a higher dig percentage is broken down to 3 factors:
- Physically keep the ball in play for longer
- Mentally frustrate opposing team offense
- Create counter attack opportunities

```python
positions = ['MB', 'OH', 'OPP', 'S']
explode = (0.05, 0.05, 0.0, 0.2)
print(hawaii_pos_kills)

# This function from here 
# https://www.geeksforgeeks.org/plot-a-pie-chart-in-python-using-matplotlib/
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} kills)".format(pct, absolute)

plt.subplots(figsize=(10, 7))
hawaii_kill_pc = plt.pie(hawaii_pos_kills,  labels=positions, autopct=lambda pct: func(pct, hawaii_pos_kills), explode=explode, startangle=0, shadow=True)
plt.title('Kill Total by Position (Hawaii)')

legend_title = f"Positions (Kill Total: {hawaii_pos_kills.sum()})"

plt.legend(positions,
          title=legend_title,
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```
